```python
import csv
import matplotlib.pyplot as plt
import numpy as np

churn = csv.reader(open('churn_data.csv'))
demographic = csv.reader(open('customer_demographics.csv'))
products = csv.reader(open('product_prices.csv'))
history = csv.reader(open('purchase_history.csv'))

# Skip headers
next(history)

# Sort history by date
HistorySorted = sorted(history, key=lambda x: x[4])

januarySum = HistorySorted[0][3]

# Determine total monthly purchases and revenue to see if there is a seasonal trend
monthlyPurchases = {
    'January': 0,
    'February': 0,
    'March': 0,
    'April': 0,
    'May': 0,
    'June': 0,
    'July': 0,
    'August': 0,
    'September': 0,
    'October': 0,
    'November': 0,
    'December': 0
}
monthlyRevenue = {
    'January': 0,
    'February': 0,
    'March': 0,
    'April': 0,
    'May': 0,
    'June': 0,
    'July': 0,
    'August': 0,
    'September': 0,
    'October': 0,
    'November': 0,
    'December': 0
}

months = {
    '01': 'January',
    '02': 'February',
    '03': 'March',
    '04': 'April',
    '05': 'May',
    '06': 'June',
    '07': 'July',
    '08': 'August',
    '09': 'September',
    '10': 'October',
    '11': 'November',
    '12': 'December'
}

for row in HistorySorted:
    data = row[4].split('-')
    month = data[1]
    monthName = months[month]
    monthlyPurchases[monthName] += 1
    monthlyRevenue[monthName] += float(row[3])

print(monthlyPurchases)
print(monthlyRevenue)

# Plot line purchases
plt.
plt.title('Monthly Purchases')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.show()


# Plot monthly revenue
plt.bar(monthlyRevenue.keys(), monthlyRevenue.values())
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```
